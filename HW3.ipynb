# **Q1:**

In order to build a model that predicts the probability that a driver will initiate an auto insurance claim in the next year. We must import, balance, and clean the dataset.

First, let's import the data and visualize the resulting table.

```python
import pandas as pd

portoFile =  '/content/portoTrain.csv'
portoData = pd.read_csv(portoFile,index_col = 'id')

portoData
```

```python
portoData.describe()
```

Next, let's see how many null values are in each column, and see if it's justifiable to either drop the column or impute it.



```python
for c in portoData.columns:
  nullCount = portoData[c].isna().sum()
  nullCount = portoData[c].value_counts().get(-1,0) + nullCount
  print(c, nullCount/len(portoData[c]))
```

As you can tell, ps_car_3_cat and ps_car_5_cat both have a significant amount of missing values, so I will drop those two. Furthermore, I will impute the remaining missing values with the mode of each column.


```python
portoData = portoData.drop(columns= ['ps_car_03_cat','ps_car_05_cat'])

#This will full out any NAN values
portoData = portoData.apply(lambda col:col.fillna(col.mode()[0]))

#The following for-loop will fill out any -1 values which are also NAN
for c in portoData.columns:
  if -1 in portoData[c].values:
    mode = portoData[c].mode()[0]
    portoData[c] = portoData[c].replace(-1,mode)

portoData.describe()
```

Next, let's check for correlation for the remaining columns and print it out on a correlation matrix. However, the matrix is rather large, so let's just try to find the correlations that are greater than 0.4 with their absolute values.

```python
corr = portoData.corr()

#The matrix is too massive to read, thus let's try to find any correlation greather than 0.4
for i in range(len(corr)):
  for j in range(i+1,len(corr)):
    if abs(corr.iloc[i,j]) >= 0.4:
      print(corr.columns[i],corr.columns[j],corr.iloc[i,j])
```

As you can tell, there are multiple columns that are heavily correlated, thus we will remove them in batches in order to ensure the dataset isn't losing too much information. First, let's look at all "ind" correlations, and get rid of "ps_ind_14", "ps_ind_06_bin", "ps_ind_15" and "ps_ind_16_bin". Afterwards, let's check the correlation again.

```python
portoData = portoData.drop(columns= ['ps_ind_14','ps_ind_06_bin',"ps_ind_15","ps_ind_16_bin"])

corr = portoData.corr()

for i in range(len(corr)):
  for j in range(i+1,len(corr)):
    if abs(corr.iloc[i,j]) >= 0.4:
      print(corr.columns[i],corr.columns[j],corr.iloc[i,j])
```

Now, let's look and get rid of the "reg" columns. From the dataset, the best columns to get rid of should be "ps_reg_02" and "ps_reg_03".

```python
portoData = portoData.drop(columns= ['ps_reg_02','ps_reg_03'])

corr = portoData.corr()

for i in range(len(corr)):
  for j in range(i+1,len(corr)):
    if abs(corr.iloc[i,j]) >= 0.4:
      print(corr.columns[i],corr.columns[j],corr.iloc[i,j])
```

Finally, let's get rid of the heavily correlated "car" columns. From the data shown above, it seems that "ps_car_12" and "ps_car_13" are the common correlated data. Thus, I will get rid of those two categories to maintain as much data as possible.

```python
portoData = portoData.drop(columns= ['ps_car_12','ps_car_13'])

corr = portoData.corr()

for i in range(len(corr)):
  for j in range(i+1,len(corr)):
    if abs(corr.iloc[i,j]) >= 0.4:
      print(corr.columns[i],corr.columns[j],corr.iloc[i,j])
```

Finally, there are no more correlated data, so we can build our models now. The models we will be building are the following:

1.   Logistic Regression
2.   Naive Bayes
3.   Random Forest
4.   Balanced Random Forest
5.   AdaBoost
6.   Gradient Boost

(Note: I will be using rebalancing techniques for the first three models: Logistic Regression, Naive Bayes, and Random Forest)

First, let's split the dataset to features and target then to training and testing.



```python
Y = portoData["target"]
X = portoData.drop(columns = ["target"])

from sklearn.model_selection import train_test_split

trainX, testX, trainY, testY = train_test_split(X,Y, test_size = 0.2, random_state = 42)
```

### **Logistic Regression with Resampling Techniques**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
from imblearn.under_sampling import RandomUnderSampler, TomekLinks
from imblearn.over_sampling import RandomOverSampler, SMOTE

scaler = StandardScaler()
trainX_scale = scaler.fit_transform(trainX)
testX_scale = scaler.fit_transform(testX)

#This is Logistic Regression without any resampling techniques
lr = LogisticRegression(random_state = 42)
print("LOGISTIC REGRESSION WITH NO RESAMPLING TECHNIQUES\n")
lr.fit(trainX_scale, trainY)
pred = lr.predict(testX_scale)

report = classification_report(testY, pred,zero_division=1 )
print("Confusion Matrix")
print(report)

probY = lr.predict_proba(testX_scale)[:, 1]
auc = roc_auc_score(testY, probY)
print("AUC Score:",auc,'\n')

#This is Logistic Regression with random undersampling
lr = LogisticRegression(random_state = 42)
print("LOGISTIC REGRESSION WITH RANDOM UNDERSAMPLING\n")
rus = RandomUnderSampler(random_state=42)
trainX_under, trainY_under = rus.fit_resample(trainX_scale,trainY)
lr.fit(trainX_under,trainY_under)
pred = lr.predict(testX_scale)

report = classification_report(testY, pred,zero_division=1 )
print("Confusion Matrix")
print(report)

probY = lr.predict_proba(testX_scale)[:, 1]
auc = roc_auc_score(testY, probY)
print("AUC Score:",auc,'\n')

#This is Logistic REgression with random oversampling
lr = LogisticRegression(random_state = 42)
print("LOGISTIC REGRESSION WITH RANDOM OVERSAMPLING\n")
ros = RandomOverSampler(random_state=42)
trainX_over, trainY_over = ros.fit_resample(trainX_scale,trainY)
lr.fit(trainX_over,trainY_over)
pred = lr.predict(testX_scale)

report = classification_report(testY, pred,zero_division=1 )
print("Confusion Matrix")
print(report)

probY = lr.predict_proba(testX_scale)[:, 1]
auc = roc_auc_score(testY, probY)
print("AUC Score:",auc,'\n')

#This is Logistic Regression with TOMEKS undersampling
lr = LogisticRegression(random_state = 42)
print("LOGISTIC REGRESSION WITH TOMEKS UNDERSAMPLING\n")
tomek = TomekLinks(sampling_strategy='auto')
trainX_tomek,trainY_tomek = tomek.fit_resample(trainX_scale,trainY)
lr.fit(trainX_tomek,trainY_tomek)
pred = lr.predict(testX_scale)

report = classification_report(testY, pred,zero_division=1 )
print("Confusion Matrix")
print(report)

probY = lr.predict_proba(testX_scale)[:, 1]
auc = roc_auc_score(testY, probY)
print("AUC Score:",auc,'\n')

#This is Logistic Regression with SMOTE oversampling
lr = LogisticRegression(random_state = 42)
print("LOGISTIC REGRESSION WITH SMOTE OVERSAMPLING\n")
smote = SMOTE(sampling_strategy='minority',random_state=42)
trainX_smote, trainY_smote = smote.fit_resample(trainX_scale,trainY)
lr.fit(trainX_smote,trainY_smote)
pred = lr.predict(testX_scale)

report = classification_report(testY, pred,zero_division=1 )
print("Confusion Matrix")
print(report)

probY = lr.predict_proba(testX_scale)[:, 1]
auc = roc_auc_score(testY, probY)
print("AUC Score:",auc,'\n')
```

### **Analysis and Conclusion**

**Comparison Between Undersampling and Oversampling**

Overall, the undersampling was consistently better than oversampling across all the models built. The undersampling consistently had an effect on the model while oversampling only sometimes had an effect. The models with no resampling techniques always overfit, and oversampling actually did not change that in some models. Since undersampling actually retains the information from the minority class, it allows for relevant data for the model to build upon. Undersampling seems to be a more reliable technique as it retains the natural essence of the dataset while oversampling does not.

**Logistic Regression Performance**

It seems that the accuracy is high with the model with no resampling, but it is obviously overfitting towards the 0 target value due to the imbalance. When you random undersample or random oversample,the recall and f1-score for the 1 target value slightly improves compared to the no resampling technique. Tomek's undersampling seems to have relatively no impact compared to the no resampling technique, and SMOTE has the worst accuracy rate compared to all the techniques. Overall, I believe random undersampling and random oversampling is the best for this dataset. My hypothesis is that there are trade-offs with all the resampling techniques even if some are more sophisticated than others. Thus, one must test all the resampling techniques to correctly model the dataset.

**Random Forest vs Balanced Random Forest**

The Balanced Random Forest seems to be better than Random Forest with no resampling. The results of the balanced random forest was similar to the random forest with random undersampling which was actually the best resampling technique for random forest. Doing some research, the balanced random forest classifier actually utilizes under sampling techniques which is why they have similar results.

# **Q2:**

Furthermore, let's try building an ensemble classifier model utilizing a VotingClassifier() and utilize the first three models in Q1 which are the following:

1.   RandomForestClassifier()
2.   LogisticRegression()
3.   NaiveBayes()

The voting classifier's mechanism combines the predictions of multiple machine learning models to produce its own prediction. By comparing all the models in the voting classifier, it chooses the best prediction depending on your soft or hard voting. (I utilized soft voting to maintain the AUC score)


```python
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from imblearn.ensemble import RUSBoostClassifier
from imblearn.ensemble import BalancedRandomForestClassifier



rfc = RandomForestClassifier()
lr = make_pipeline(StandardScaler(),LogisticRegression(max_iter = 1000))
nb = GaussianNB()
rus = RUSBoostClassifier()
gradient = GradientBoostingClassifier()
brfc = BalancedRandomForestClassifier()

print("VOTING CLASSIFIER WITH RANDOM FOREST, LOGISTIC REGRESSION, NAIVE BAYES (VOTING = SOFT)")
voting = VotingClassifier(estimators=[
                                      ('random_forest', rfc),
                                      ('logistic_regression', lr),
                                      ('naive_bayes', nb),
                                      ('rus', rus),
                                      ('gradient',gradient),
                                      ('balanced_rfc',brfc)], voting = 'soft',n_jobs = -1)
voting.fit(trainX,trainY)
pred = voting.predict(testX)
report = classification_report(testY, pred,zero_division=1 )
print("Confusion Matrix")
print(report)

probY = voting.predict_proba(testX)[:, 1]
auc = roc_auc_score(testY, probY)
print("AUC Score:",auc,'\n')
```

### **Analysis and Conclusion**


**Comparison with Adaboost and Gradient Boost**

Overall, the results of the voting classifier was similar to the gradient boost classifier. However, it seems to be overfitting as both of those models have an accuracy rate of 96-97%. Adaboost seems to be better at taking into account the minority class even though the accuracy score drops significantly.

**Ensemble Model Performances**

For the voting classifier specifically, you have to pick your models carefully. You need to ensure that the results of each model has a level of fairness and diversity for the voting to be balanced. For example, you don't want multiple models overfitting as it essentially will lead the voting classifier to overfit.

For other ensemble models, it mainly depends on the dataset and what the objective is, and just need to be tested to find which model suits best.

Overall, the voting classifier is difficult as one must study the dataset AND the models within the voting classifier.
